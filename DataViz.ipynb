```python
import pandas as pd
import numpy as np
```

```python
incidents = pd.read_csv("incident.csv")
```

```python
df = pd.read_csv('level1_data.csv')
```

```python
incidents
```

```python
df = df.drop_duplicates("Incident ID")
df
```

```python
incidents["Incident ID"] = incidents["Incident ID"].astype(int)
```

```python
incidents
```

```python
df = pd.merge(left=incidents, right=df, how="left", left_on='Incident ID',right_on='Incident ID')
```

```python
cols = df.columns[4:101]
df1 = df[cols]
```

```python
df1 = df.groupby(by='State').sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = list(df1.sum().sort_values(ascending = False).index[10:30])
cols.sort()
rows = list(df1.T.sum().sort_values(ascending = False).head(15).index)
df_heat = df1[cols]
df_heat = df_heat[df_heat.index.isin(rows)]
```

```python
sns.set()

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(df_heat)
```
